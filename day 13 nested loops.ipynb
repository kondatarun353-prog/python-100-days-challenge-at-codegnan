# Nested loops

```python
for i in range(1,3):
    for j in range(1,11):
        print(f"{i}*{j}={i*j}")
    print("_________")
```

```python
for i in range(0,11):
    print(i,end=" ")
```

```python
#print 3 table in functions
def three_table(n):
    for i in range(1,11):
        print(f"{n} * {i} = {i*n}")
three_table(3)
```

```python
#rectangular pattern
n=int(input())
for i in range(n):
    for j in range(1,n+1):
        print("*",end=" ")
    print()
    print()
```

```python
for i in range(6):
    for j in range(6):
        if i==j:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if i==j or j==n-1-i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```
