```python
#who is the topper among 10 students (print name #average marks of total students
#each subject average marks
print("Marks details of students")
n=int(input("Enter number of students:"))
names=[]
marks=[]
class_marks=[]
percentage=[]
print("Enter all student names")
for i in range(n):
    print(f"\nEnter the roll number {i+1} student marks")
    name=input("Enter the student name:")
    print("Enter Telugu,Hindi,English,maths,Science,Social marks respectively")
    T,H,E,M,S,S=map(int,input("Enter the student marks").split()[:6])
    student_allsub_marks=T+H+E+M+S+S
    student_percentage=(student_allsub_marks/600)*100
#store all values in list
    names.append(name)
    marks.append([T,H,E,M,S,S])
    class_marks.append(student_allsub_marks)
    percentage.append(student_percentage)
#print the topper among them
    print("\nTopper in the class is")
    top_marks=max(class_marks)
    topper_position=[]
for i in range(n):
    if class_marks[i]==top_marks:
        topper_position.append(i)
#to print each marks student
    for i in topper_position:
        print("name of the student is: ",name[i])
        print("marks of student is:",marks[i])
        print("Total marks is",class_marks[i])
        print("Percentage",percentage[i])
        print("All student details:")
    for i in range(n):
        print(
            "Name:",name[i],
            "| Marks:",marks[i],
            "| total marks:", class_marks[i],
            "| percentage:",percentage[i])
        
    
    
```

```python
#find the smallest value in the list without using build in functions
#find the largest values in the list
#find the length of list
def find_largest(lst):
    if not lst:
        return "error:list cannot be empty"
    largest=lst[0]
    for item in lst[1:]:
        if item>largest:
            largest=item
    return largest
numbers=[10,20,30,90]
result=find_largest(numbers)
print(result)
```

```python
#smallest
def find_smallest(lst):
    if not lst:
        return"error:list cannot be empty"
    smallest=lst[0]
    for item in lst[1:]:
        if item<smallest:
            smallest=item
        return smallest
numbers=list(map(int,input().split()))
result=find_smallest(numbers)
print(result)

        
    
```
